```python
def frequency(data):
    freq = data[['ProductDescription','Job_Count']]
    x = freq.groupby('ProductDescription').count()
    y = x.sort_values(['Job_Count'], ascending=[False])
    t = y.head(50)
    print(t)
```

```python
def frequency_month(data):
    freq = data[['ProductDescription','Job_Count', 'DeliveryDate']]
    
    x = freq.groupby('ProductDescription').count()
    y = x.sort_values(['Job_Count'], ascending=[False])
    t = y.head(50)
    print(t)
```
